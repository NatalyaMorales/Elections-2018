# Data Frame DIPUTACIÓN en CSV

```python
ruta_csv = '/content/drive/MyDrive/Colab Notebooks/Minería/Lab1/2018_SEE_DIP_FED_MR_NAL_SEC.csv'
df = pd.read_csv(ruta_csv, dtype = {'ID_ESTADO': str, 'ID_DISTRITO': str, 'ID_MUNICIPIO': str})
df['ID_ESTADO'] = df['ID_ESTADO'].astype(str)
df['ID_DISTRITO'] = df['ID_DISTRITO'].astype(str)
#df['ID_MUNICIPIO'] = df['ID_MUNICIPIO'].astype(str)
```

```python
df
```

```python
 # Funciones para agregar ceros
def Corregir(x):
  if len(x) == 1:
    return '0' + x
  if len(x) == 2:
    return x

def Corregir2(x):
  if len(x) == 1:
    return '00' + x
  if len(x) == 2:
    return '0' + x
  if len(x) == 3:
    return x
```

```python
# Se corrigen los ceros a la derecha
df['ID_ESTADO'] = df['ID_ESTADO'].apply(Corregir)
df['ID_DISTRITO'] = df['ID_DISTRITO'].apply(Corregir2)
df['ID_MUNICIPIO'] = df['ID_MUNICIPIO'].apply(Corregir2)
```

```python
# Se combinan las columnas 'ID_ESTADO' y 'ID_DISTRITO' en la columna 'index'
df['index'] = df['ID_ESTADO'] + df['ID_DISTRITO']
# Se combinan las columnas 'ID_ESTADO' y 'ID_MUNICIPIO' en la columna 'index_mun'
df['index_mun'] = df['ID_ESTADO'] + df['ID_MUNICIPIO']
```

```python
# Se imprime la columna 'index'
df['index']
```

```python
# Se imprime df
df
```

```python
# Filtra df para las filas que tengan en 'NOMBRE_ESTADO' a 'PUEBLA'
df_filtrado = df[df['NOMBRE_ESTADO'] == 'PUEBLA']
df_filtrado
```

```python
# Se agrupan los datos de acuerdo al municipio
df_agrupado = df_filtrado.groupby(by='index_mun')
df_agrupado

# Se suman los datos
suma = df_agrupado.sum()
suma

# Se filtra la suma por las columnas seleccionadas
suma_filtrado = suma[['PAN', 'PRI', 'PRD', 'PVEM',
       'PT', 'MC', 'NA', 'MORENA', 'ES', 'PAN_PRD_MC', 'PAN_PRD', 'PAN_MC',
       'PRD_MC', 'PRI_PVEM_NA', 'PRI_PVEM', 'PRI_NA', 'PVEM_NA',
       'PT_MORENA_ES', 'PT_MORENA', 'PT_ES', 'MORENA_ES', 'CAND_IND1',
       'CAND_IND2', 'NUM_VOTOS_CAN_NREG', 'NUM_VOTOS_NULOS', 'TOTAL_VOTOS',
       'LISTA_NOMINAL']]
suma_filtrado
```
